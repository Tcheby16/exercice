# EXERCICE 1 : Mise en place d'un calendrier 

```python
def afficher_calendrier(n_jours, premier_jour):
    """
    Affichage d'un calendrier correctement aligné au format demandé
    :param n_jours: Nombre de jours dans le mois
    :param premier_jour: Jour de la semaine où commence le mois (1 pour lundi, ..., 7 pour dimanche)
    """
    jours_semaine = ["LUN", "MAR", "MER", "JEU", "VEN", "SAM", "DIM"]
    print(" ".join(jours_semaine))

    # Ajustement pour le décalage du premier jour
    espace_initial = (premier_jour - 1) * 4  # 4 espaces pour chaque jour
    print(" " * espace_initial, end="")

    # Génération des jours du mois
    for jour in range(1, n_jours + 1):
        print(f"{jour:>3} ", end="")  # Alignement des jours 
        
        # Retour à la ligne après dimanche (chaque 7 jours complets)
        if (jour + premier_jour - 1) % 7 == 0:
            print()  # Saut de ligne pour la nouvelle semaine
    print()  # Ligne vide après le calendrier


# Entrée des données par l'utilisateur
n_jours = int(input("Entrez le nombre de jours dans le mois : "))
premier_jour = int(input("Entrez le numéro du jour démarrant le mois (1 pour lundi, ..., 7 pour dimanche) : "))

# Validation des entrées
if 28 <= n_jours <= 31 and 1 <= premier_jour <= 7:
    print() # Ligne vide pour une bonne présentation
    afficher_calendrier(n_jours, premier_jour)
else:
    print() # Ligne vide pour une bonne présentation
    print("Données invalides. Assurez-vous que le nombre de jours dans le mois est entre 28 et 31, et que le jour de départ est entre 1 et 7.")
```

# EXERCICE 2 : Mise en place d'un jeu sur les régions et leurs chefs-lieux

```python
import random

# Base de données des régions et de leurs chefs-lieux
regions = {
    "Agneby Tiassa": "Agboville",
    "Bafing": "Touba",
    "Bagoue": "Boundiali",
    "Belier": "Toumodi",
    "Bere": "Mankono",
    "Bounkani": "Bouna",
    "Cavally": "Guiglo",
    "District Autonome Abidjan": "Abidjan",
    "District Autonome Yamoussoukro": "Yamoussoukro",
    "Folon": "Minignan",
    "Gbokle": "Sassandra",
    "Gbeleke": "Bouaké",
    "Goh": "Gagnoa",
    "Gontougo": "Bondoukou",
    "Grands Ponts": "Dabou",
    "Guemon": "Duékoué",
    "Hambol": "Katiola",
    "Haut Sassandra": "Daloa",
    "Ifou": "Daoukro",
    "Indénie Djuablin": "Abengourou",
    "Kabadougou": "Odienné",
    "La Mé": "Adzopé",
    "Loh Djiboua": "Divo",
    "Marahoué": "Bouaflé",
    "Nawa": "Soubré",
    "N'zi": "Dimbokro",
    "Poro": "Korhogo",
    "San-Pedro": "San-pédro",
    "Sud Comoé": "Aboisso",
    "Tchologo": "Ferkessédougou",
    "Tonkpi": "Man",
    "Worodougou": "Séguéla"
}

# Liste des meilleurs scores (pseudo, score)
meilleurs_scores = [("Charly", 400), ("Trésor", 350), ("Marius", 300), ("Kalou", 250), ("Karim", 200)]

def afficher_meilleurs_scores():
    """Affiche les 5 meilleurs scores"""
    print("\n=== TOP 5 DES MEILLEURS SCORES ===")
    for i, (pseudo, score) in enumerate(meilleurs_scores, start=1):
        print(f"{i}. {pseudo} - {score} points")
    print("=================================\n")

def jouer():
    """Fonction principale pour une partie de jeu"""
    print("\nBienvenue au jeu des régions de la Côte d'Ivoire !")
    afficher_meilleurs_scores()

    # Initialisation
    pseudo = input("Entrez votre pseudo : ")
    questions_restantes = list(regions.keys())
    score = 0
    bonnes_reponses = 0

    # Boucle pour 10 questions
    for _ in range(10):
        # Choisir une région aléatoire
        region = random.choice(questions_restantes)
        chef_lieu = regions[region]
        
        # Poser la question
        reponse = input(f"Quel est le chef-lieu de la région {region}? ").strip()
        
        # Vérifier la réponse
        if reponse.lower() == chef_lieu.lower():
            print("Bonne réponse !")
            score += 50  # Par exemple, 50 points par bonne réponse
            bonnes_reponses += 1
        else:
            print(f"Réponse incorrecte ! Le chef-lieu de {region} est {chef_lieu}.")
        
        # Retirer la région des questions restantes
        questions_restantes.remove(region)
    
    # Afficher le score final
    print(f"\nFin du jeu, {pseudo} ! Vous avez {bonnes_reponses} réponses correctes sur 10.")
    print(f"Votre score est de : {score} points.")
    
    # Mise à jour des meilleurs scores
    meilleurs_scores.append((pseudo, score))
    meilleurs_scores.sort(key=lambda x: x[1], reverse=True)
    if len(meilleurs_scores) > 5:
        meilleurs_scores.pop()  # Garder uniquement les 5 meilleurs
    
    afficher_meilleurs_scores()

def main():
    """Fonction principale pour gérer le jeu"""
    while True:
        jouer()
        rejouer = input("Voulez-vous effectuer une autre partie ? [oui/non] : ").strip().lower()
        if rejouer != 'oui':
            print("Merci d'avoir joué ! À bientôt.")
            break

# Lancer le jeu    
if __name__ == "__main__":
    main()
```
